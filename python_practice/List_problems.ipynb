- Problem: Given a list of words, find the word with the maximum length.

```python
fruits=['apple','banana','orange','watermelon','peach','graps'];
```

```python
fruits
```

```python
word=fruits[0];
for fruit in fruits:
    if len(fruit)>len(word):
        word=fruit;
print(word)
```

- Problem: Given a list of integers, find the sum of all elements at even indices.

```python
import random;
numbers=[random.randint(1,100) for i in range(25)];
```

```python
numbers
```

```python
len(numbers)
```

```python
total_sum=0;
for index,value in enumerate(numbers):
    if index%2==0:
        total_sum+=value;
print(total_sum);
```

- Problem: Given a list of strings, create a new list containing only the strings that start with a vowel (a, e, i, o, u) ignoring the case.

```python
fruits=['apple','banana','Orange','watermelon','Peach','graps'];
```

```python
new_fruits=[];
vowels=['A','a','E','e','I','i','O','o','U','u'];
for i in fruits:
    if i[0] in vowels:
        new_fruits.append(i)
```

```python
new_fruits
```

simple

```python
new_fruits=[word for word in fruits if word[0].lower() in vowels];
```
