### Discretization and quantiling

### Discretization:

- Discretization is the process of turning continuous values into categories or bins. It's like putting these values into buckets. Imagine you have the heights of people, and you want to categorize them into "short," "average," and "tall" groups. pd.cut(). 

### Quantiling
- it uses the distribution of the data. It divides the data into equal parts, making sure each part has an equal number of values. pd.qcut()

```python
heights = [150, 165, 175, 180, 160, 185, 170, 155];
```

```python
pd.cut(heights, 3)
```

```python
bin_edges = [149,164,179,194];
pd.cut(heights, bins=bin_edges)
```

```python
pd.cut(heights, bins=bin_edges, labels=['short','average','tall'])
```

```python
arr = np.random.randn(20);
arr
```

```python
pd.cut(arr, 4)
```

#### Quantiling

```python
heights = [150, 165, 175, 180, 160, 185, 170, 155];
```

```python
pd.qcut(heights, 3)
```

```python
q = [0, 0.333333, 0.66666667, 1];
pd.qcut(heights, q)
```

```python
pd.qcut(heights, q, labels=['short', 'average', 'tall'])
```

```python
pd.qcut(heights, [0, 0.5, 1.0], labels=['short','tall'])
```

### Function Application
- 1. Tablewise Function Application: pipe()
- 2. Row or Column-wise Function Application: apply()
- 3. Aggregation API: agg() and transform()
- 4. Applying Elementwise Functions: applymap()

#### 1. Tablewise Function Application: pipe()

```python
def extract_city_name(df):
    df['city_name'] = df['city_and_code'].str.split(',').str.get(0);
    return df;

def add_country_name(df, country_name=None):
    df['city_and_country'] = df['city_name']+" "+country_name;
    return df;
```

```python
df_p = pd.DataFrame({"city_and_code": ["Chicago, IL"]})
```

```python
add_country_name(extract_city_name(df_p), country_name='US')   ### Style 1
```

#### pandas encourages the second style, which is known as method chaining. pipe makes it easy to use your own or another librarys functions in method chains, alongside pandas methods.


```python
## Second Style
df_p.pipe(extract_city_name).pipe(add_country_name, country_name='US')
```

```python
df_p = pd.DataFrame({"city_and_code": ["Chicago, IL","Islamabad, IS"]})
```

```python
df_p
```

```python
df_p.pipe(extract_city_name).pipe(add_country_name, country_name="pakistan")
```

 -  iterrows(): Iterate over the rows of a DataFrame as (index, Series) pairs. This converts the rows to Series objects, which can change the dtypes and has some performance implications.
 -  itertuples(): Iterate over the rows of a DataFrame as namedtuples of the values. This is a lot faster than iterrows(), and is in most cases preferable to use to iterate over the values of a DataFrame.

```python
for index, row in df.iterrows():
    print(row)
```

```python
for index, row in df.iterrows():
    print(row, '\n')
```

```python
for t in df.itertuples():
    print(t)
```

```python
for i,c1,c2 in df.itertuples():
    print(c1,c2)
```

```python
for i,v in {'a':[34,53],'b':[654,23]}.items():
    print(i,v)
```

```python
for i,v in df.items():
    print(i,v, '\n')
```

```python
for i,v in df.items():
    print(i, np.array(v).round(2));
```

```python
df
```

```python
for i,row in df.iterrows():
    print(i)
    print(row)
```

```python
df1 = pd.DataFrame([[12,4.4]], columns=['int', 'floatt']);
df1
```
