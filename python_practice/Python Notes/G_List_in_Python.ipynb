#### enumerate()

- In Python, the enumerate() function is a built-in function that allows you to loop over an iterable (such as a list, tuple, string, or any other iterable) while keeping track of both the index and the corresponding item in the iterable. It returns an iterator of tuples containing the index and the element from the original iterable.

```python
L1
```

```python
for index,item in enumerate(L1):
    print(index,item)
```
