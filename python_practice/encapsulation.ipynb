### Encapsulation in python

```python
class Atm:
    
    def __init__(self):
        self.__pin='';      # private           => _Atm__pin  Python actually conver variable with __ to this name
        self._balance=0;    # protected         => _Atm_balance
        
    def get_pin(self):
        print("the pin is : ",self.__pin);
        
    def set_pin(self,p):
        if type(p)==str:
            self.__pin=p;
        else:
            print("Not Allowed");
```

```python
HBL=Atm();
```

```python
HBL.__pin='233k';  # this will create a new variable with the name of pin
```

```python
print("ID of HBL.__pin variable is {} which is of type {}".format(id(HBL.__pin),type(HBL.__pin)));
```

```python
HBL.set_pin('khan123');
```

```python
HBL.get_pin();
```

```python
print(HBL.__pin);  # the HBL.__pin is still 233k as we have changed the value of  __pin variable of class using set_pin() method. 
```

#### Showing that Classes are Mutable

```python
print("ID Of HBL object before calling the set_pin method",id(HBL));
```

```python
HBL.set_pin("Khan");
```

```python
HBL.get_pin();
```

```python
print("ID of HBL object after calling the set_pin method",id(HBL));
```

```python
# since the id of HBL is same before and after the calling of set_pin method so it shows that classes are mutable 
```
