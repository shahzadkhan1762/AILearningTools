### ATM Class

```python
class ATM:
    
    def __init__(self):
        self.pin='';
        self.balance=0;
        self.menu();
        
    def menu(self):
        option=int(input("""
            (1) : Enter 1 to set PIN
            (2) : Enter 2 to Deposit Money
            (3) : Enter 3 to Withdrow Money
            (4) : Enter 4 to check balance
            (5) : Enter 5 to Exit
        """));
        
        if option==1:
            self.set_pin();
        elif option==2:
            self.deposit();
        elif option==3:
            self.withdraw();
        elif option==4:
            self.check_balance();
        elif option==5:
            exit(0);
        else:
            print("Please Enter a valid option");
            self.menu();
    
    def set_pin(self):
        self.pin=input("Enter pin for your account : ");
        print("PIN Added successfully!");
        self.menu();
        
    
    def deposit(self):
        tpin=input("Enter your pin : ");
        if tpin==self.pin:
            amount=int(input("Enter amount you wanna deposit : "));
            self.balance+=amount;
            print("Deposit Seccessful!");
        else:
            print("Sorry! your pin is incorrect");
        self.menu();
        
            
    def withdraw(self):
        tpin=input("Enter your pin : ");
        if tpin==self.pin:
            amount=int(input("Enter amount you wanna withdraw : "));
            if amount<=self.balance:
                self.balance-=amount;
                print("Amound Withdrawed!")
            else:
                print(f"You don't have {amount} amount in your account");
                print("your current balance is : ",self.balance);
        else:
            print("Sorry! your pin is incorrect");
            
        self.menu();
        
    
    def check_balance(self):
        tpin=input("Enter your pin : ");
        if tpin==self.pin:
            print("Your current balance is : ",self.balance);
        else:
            print("Sorry! your pin is incorrect");
        self.menu();
```
