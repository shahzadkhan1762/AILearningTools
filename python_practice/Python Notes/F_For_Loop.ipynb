### for loop

- A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

#### Syntax

```python
#    for i in range(3):
#        statements
```

Example

```python
for i in range(10):
    print(i)
```

```python
L=[6,2,6,23,67,34];

for item in L:
    print(item);
```

```python
T=('apple','mango','banana');

for item in T:
    print(item);
```

#### Nested For Loop

```python
row=int(input("Enter number of rows : "));
for i in range(1,row+1):
    for j in range(1,i):
        print('*',end=' ');
    print();
```

### while loop

- With the while loop we can execute a set of statements as long as a condition is true.

#### Syntax

```python
#  While condition:
#      Statement(s);
```

Example

```python
a=5; 

while a<10:
    print('pakistan');
    a+=1;
```
