#### Working with student_record text file in text_file directory

```python
s_file_path="text_files\student_record.txt";
with open(s_file_path) as file_object:
    lines=file_object.readlines();
```

```python
print(lines);
```

```python
columns=list(lines[0].strip().split(','));
```

```python
rows=[line.strip().split(',') for line in lines[1:]]
```

```python
rows
```

```python
students=dict(zip(columns,rows[0]));
```

```python
students
```

```python
students=[dict(zip(columns,rows[i])) for i in range(len(rows))];
```

```python
students
```

#### convert age and marks to int

```python
for student in students:
    for key,value in student.items():
        if key=='age':
            student[key]=int(value);
        if key=='marks':
            student[key]=int(value);
```
