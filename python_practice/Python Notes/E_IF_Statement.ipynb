### if Statement

- if statement execute a block of code based on a given condition.

#### Syntax:

```python
# if condition:
#     statement(s);
```

#### Example

```python
s_email='khan@gmail.com';
s_password='khan123';

email=input("Enter Email : ");
password=input("Enter Password : ");

if email==s_email and password==s_password:
    print("Login Successful!");
```

### if else Statement

- if else statement exucute a block of code if a given condition is true, else other block of code is exucuted. 

```python
# if condition:
#    statement(s);
# else:
#    statement(s);
```

```python
s_email='khan@gmail.com';
s_password='khan123';

email=input("Enter Email : ");
password=input("Enter Password : ");

if email==s_email and password==s_password:
    print("Login Successful!");
else:
    print("Sorry! your Password/email is incorrect!");
```
