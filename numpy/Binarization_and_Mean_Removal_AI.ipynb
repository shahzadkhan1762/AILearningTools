#### Binarization
- This process is used when we want to convert our numerical values into boolean values.
Let's use an inbuilt method to binarize input data using 2.1 as the threshold value.

```python
binarize_data=preprocessing.Binarizer(threshold=0.5).transform(input_data);
```

```python
binarize_data
```

#### Mean Removal

```python
input_data=np.random.randint(low=0, high=100, size=(3,6));
```

```python
input_data
```

```python
input_data.mean(axis=0)
```

```python
input_data.std(axis=0)
```

```python
# Remove mean
# scale function perform the following operation on each element in the input array
# scaled_X = (X - mean(X)) / std(X)  

scale_data=preprocessing.scale(input_data);
```

```python
scale_data
```

```python
scale_data.mean(axis=0)
```

```python
scale_data.std(axis=0)
```

```python
#scaled_X = (X - mean(X)) / std(X)

(input_data[0,0]-input_data.mean(axis=0)[0])/input_data.std(axis=0)[0]
```
