### Aggregation

```python
class Customer:
    
    def __init__(self):
        self.name=None;
        self.gender=None;
        self.age=None;
        self.address=Address();
    
    def add_info(self,name,gender,age,city,pincode,state):
        self.name=name;
        self.gender=gender;
        self.age=age
        self.address.set_address(city,pincode,state);
        
    def show_info(self):
        print("Name : ",self.name);
        print("Gender : ",self.gender);
        print("Age : ",self.age);
        self.address.show_address();
```

```python
class Address:
    
    def __init__(self):
        self.city='';
        self.pincode=None;
        self.state='';
    
    def set_address(self,city,pincode,state):
        self.city=city;
        self.pincode=pincode;
        self.state=state;
    
    def show_address(self):
        print("Address : ",self.city,self.pincode,self.state);
```

```python
customer=Customer();
```

```python
customer.show_info();
```

```python
customer.add_info("abbas",'Male',20,'Hangu',25000,'kp');
```

```python
customer.show_info();
```
