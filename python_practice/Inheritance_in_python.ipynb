### Inderitance

```python
class User:
    
    def login(self):
        print("Login");
        
    def register(self):
        print("Resgister");
```

```python
class Student(User):
    
    def enroll(self):
        print("Enroll");
    
    def review(self):
        print("Review");
```

```python
s=Student();
```

```python
s.login();
```

```python
s.register();
```

```python
s.enroll();
```

```python
s.review();
```

### IF Child Class has no Constructor

```python
class Phone:
    
    def __init__(self,price,brand,camera):
        self.price=price;
        self.brand=brand;
        self.camera=camera;
        
    def show(self):
        print("Price : ",self.price);
        print("Brand : ",self.brand);
        print("Camera : ",self.camera);
```

```python
class SmartPhone(Phone):
    pass;
```

```python
smartPhone=SmartPhone(20000,'IPhone',14);
```

```python
smartPhone.show();
```
