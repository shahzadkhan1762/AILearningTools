### Investment Class

```python
class Investment:
    def __init__(self, principal, interest):
        self.principal = principal;
        self.interest = interest;
    
    def __str__(self):
        return f"Principal - ${self.principal}.00,  Interest rate - {self.interest}";
        
    def value_after(self, n):
        return (self.principal * ((1 + self.interest)**n));
    
invest = Investment(5000000, 10);
print(f"The value of investment after 5 years is : {invest.value_after(5)}");
print(invest)
```
