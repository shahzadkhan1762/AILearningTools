#### Creating Array using different methods

```python
import numpy as np;
```

***This section covers np.array(), np.zeros(), np.ones(), np.empty(), np.arange(), np.linspace(), dtype***

```python
npArray=np.array([[234,623,623],[634,723,136]],np.int16);
```

```python
print(npArray);
```

```python
for i in range(2):
    for j in range(3):
        print(npArray[i][j]);
        j=j+1;
    i=i+1;
```

```python
npArray.shape
```

```python
npArray.size
```

```python
npArray.dtype
```
