## Sets

- A Python set is the collection of the unordered items.

### Four Rules For Sets

- (1) : Sets do not allow duplicates.
- (2) : Sets have no indexing/slicing.
- (3) : Set don't allow mutable data types.
- (4) : Sets iteself are mutable data types.
