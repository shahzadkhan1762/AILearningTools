**Word Frequency Counter: Given a list of words, create a dictionary where the keys are the unique words and the values are the frequencies of each word in the list.**

```python
wordFrequency={'apple':4,'banana':3,'strawberry':2,'orange':8};
```

```python
for key in wordFrequency.keys():
    print(key);
```

```python
for key,value in wordFrequency.items():
    print("Fruite name is : "+key,end='   ');
    print("Frequency of "+key+" are : "+str(value));
```

**Student Database:**

```python
studentDB={1:{'name':'khalid','age':20,'grade':89},2:{'name':'waqas','age':23,'grade':90},3:{'name':'shabeer','age':40,'grade':60}};
```

```python
def std_info():
    for key,value in studentDB.items():
        print("\nID : ",key);
        for k,v in value.items():
            print(k," : ",v);
```

```python
std_info();
```

```python
def avg_grade():
    total_grade=0;
    count=0;
    for value in studentDB.values():
        total_grade += value['grade'];
        count+=1;
    average=total_grade/count;
    return average;
            
```

```python
print(avg_grade());
```
