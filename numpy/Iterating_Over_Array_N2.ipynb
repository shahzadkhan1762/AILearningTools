### Tracking an Index or Multi-Index

```python
arr
```

```python
it=np.nditer(arr, flags=['multi_index']);
while not it.finished:
    print("Index : %s   value : %d" % (it.multi_index, it[0]));
    it.iternext();
```

```python
oneDarr=np.arange(20);
```

```python
oneDarr
```

```python
it=np.nditer(oneDarr, flags=['f_index']);
for i in it:
    print("Index : <%d>   value : %d" % (it.index, it[0])); # not need of it.iternext() because for loop automatically call the next functio for it.
```

```python
arr=np.arange(24).reshape(4,6);
arr
```

```python
# For multi Dimensional array

it=np.nditer(arr, flags=['f_index']);
for i in it:
    print("Index : <%d>   value : %d" % (it.index, it[0]));   # not need of it.iternext() because for loop automatically call the next functio for it.
```

```python
arr
```

```python
iterator=np.nditer(arr, flags=['multi_index']);
while not iterator.finished:
    print("Value : %d   Index : %s" % (iterator[0], iterator.multi_index));
    iterator.iternext();
```

### Iterating as a Specific Data Type

```python
ar=np.arange(12);
for i in np.nditer(ar, flags=['external_loop'], op_flags=['readonly','copy'], op_dtypes=[np.int64]):
    print(i);
print(ar.dtype)
```

```python
for i in np.nditer(ar, flags=['external_loop'], op_flags=['readonly', 'copy'], op_dtypes=[np.complex128]):
    print(i)
```

```python
for i in np.nditer(ar, flags=['external_loop', 'buffered'], op_dtypes=[np.float64]):
    print(i)
```

```python
for i in np.nditer(ar, flags=['external_loop', 'buffered'], op_dtypes=[np.int32]):
    print(i)
```

```python
for i in np.nditer(ar, flags=['external_loop', 'buffered'], op_dtypes=[np.int16], casting='same_kind'):
    print(i);
```

```python
a=np.array([i for i in np.nditer(ar, flags=['external_loop', 'buffered'], op_dtypes=[np.int16], casting='same_kind')]);
```

```python
array=np.arange(12).reshape(2,6);
for i in np.nditer(array, flags=['buffered'], op_dtypes=[np.float64]):
    print(i, end=' ');
```

```python
def ftn():
    array=np.arange(12).reshape(2,6);
    for i in np.nditer(array, flags=['external_loop'], op_flags=['readonly', 'copy'], op_dtypes=[np.float64]):
        print(i**i, end=' ');
    print();
```

```python
%time ftn();
```

```python
def ftn1():
    array=np.arange(12).reshape(2,6);
    for i in np.nditer(array):
        print(np.float64(i**i), end=' ');
    print();
```

```python
%time ftn1();
```
