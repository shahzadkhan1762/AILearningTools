```python
import numpy as np;
```

```python
arr = np.ones((3,5), dtype='h', order='C');

for item in np.nditer(arr):
    print(item, end=' ')
```

```python
arr = np.arange(2,5,0.1, dtype='e').reshape(5,6);
```

```python
arr
```

```python
for item in np.nditer(arr):
    print(item, end=' ')
```

```python
arr = arr.copy(order='F')
```

```python
arr
```

```python
for item in np.nditer(arr):
    print(item, end=' ')
```

```python
arr
```

```python
for item in np.nditer(arr, order='K'):    # Because the array is stored in F order
    print(item, end=' ')
```

```python
for item in np.nditer(arr, order='C'):
    print(item, end=' ')
```

```python
# op_flags
# Modifying the element of array

for item in np.nditer(arr):
    item[...]=item+1;
```

```python
for item in np.nditer(arr, op_flags=['readwrite']):
    item[...]=item+1;
```

```python
arr
```

```python
# Using an External Loop

arr
```

```python
for item in np.nditer(arr, flags=['external_loop']):
    print(item, end=' ')
```

```python
for item in np.nditer(arr, flags=['external_loop'], order='F'):
    print(item, end=' ')
```

```python
for item in np.nditer(arr, flags=['external_loop'], order='C'):
    print(item, end=' ')
```

```python
arr
```

```python
# Tracking an Index or Multi-Index

it = np.nditer(arr, flags=['f_index']);

for item in it:
    print('Index %d    value %d'%(it.index, item));
```

```python
it = np.nditer(arr, flags=['multi_index']);
while not it.finished:
    print('index %s    value %d'%(it.multi_index,  it[0]));
    it.iternext()
```

```python
arr1 = np.arange(3,9,1).reshape(2,3);
```

```python
arr1
```

```python
iterator = np.nditer(arr1, flags=['multi_index'], op_flags=['readwrite']);

while not iterator.finished:
    print(iterator.multi_index[0]);
    iterator.iternext();
```

```python
iterator = np.nditer(arr1, flags=['multi_index'], op_flags=['readwrite']);

while not iterator.finished:
    print('Index {}   Value {}'.format(iterator.multi_index, iterator[0]))
    iterator.iternext();
```

```python
arr1
```

```python
iterator = np.nditer(arr1, flags=['multi_index'], op_flags=['readwrite']);

while not iterator.finished:
    iterator[0]=iterator.multi_index[1]-iterator.multi_index[0]
    iterator.iternext();
```
