### Basic Data types

#### Integer Data Type

- Integers are numbers without fractional part. The size of Integer data type in python is 4Bytes.

```python
a=23;
print(a);
```

```python
type(a)
```

```python
print(2+2)
```

#### Float Data Type

- Python calls any number with a decimal point a float. And the size of float Numbers are 64bit or 8Bytes.

```python
a=5.4;
b=4.6;
print(a+b);
```

```python
print(type(a));
```

#### String Data Type

```python
print('Pakistan');
```
