#### Writing to and Empty file

```python
file_path='programming.txt';
with open(file_path,'w') as file_object:
    file_object.write('I Love Programming!');
```

```python
file_path='text_files\programming.txt';
with open(file_path,'w') as file_object:
    file_object.write('I Love programming');
```

#### Two arguments in open method

**Argument 1: path of the file that you wanna open**

**Argument 2: Mode in which you wanna open the file**

***'r' : Reading mode***

***'w' : Writing mode***

***'r+' : Reading + Writing mode***

***'a' : Append mode***

#### Writing Mutiple lines

```python
with open(file_path,'w') as file_object:
    file_object.write("I Love Programming!\n");
    file_object.write("I Love Creating New Games!\n");
```

#### Appending to a File

```python
with open(file_path,'a') as file_object:
    file_object.write("This is my first day in file handling in python\n");
    file_object.write("I will learn more about file latter\n");
```
