### Instance as an Attribute

```python
class User():
    
    def __init__(self, first_name='N/A', last_name='N/A', age='N/A', address='N/A', password='N/A'):
        self.first_name=first_name;
        self.last_name=last_name;
        self.age=age;
        self.address=address;
        self.password=password;
        self.login_attempts=0;
        
    def desc_user(self):
        print("Detail of User");
        print("First Name : "+self.first_name);
        print("Last Name : "+self.last_name);
        print("Age : ",self.age);
        print("Address : "+self.address);
        print("Password : "+self.password);
        
    def great_user(self):
        print("welcome "+self.first_name+" "+self.last_name+"!");
        
    def increment_login_attempts(self):
        self.login_attempts+=1;
        
    def reset_login_attempts(self):
        self.login_attempts=0;
```

```python
class Previliges():
    
    def __init__(self):
        self.previliges=["can add post", "can delete post", "can ban user"];
    
    def show_previliges(self):
        print("List of previliges with current user");
        for i in self.previliges:
            print(i);    
```

```python
class Admin(User):
    
    def __init__(self, first_name, last_name, age, address, password):
        super().__init__(first_name,last_name,age,address,password);
        self.previliges=Previliges();
```

```python
admin=Admin('Abbas','Khan',20,'hangu','dfhddfhd');
```

```python
admin.desc_user();
```

```python
admin.previliges.show_previliges();
```

```python
admin.great_user();
```
