```python
def mul(a,b):
    if b==1:
        return a;
    else:
        return a+mul(a,b-1);
```

```python
print(mul(4,5));
```

```python
def fact(number):
    if number==1 or number==0:
        return 1;
    else:
        return number*fact(number-1);
```

```python
fact(1)
```

```python
fact(2)
```

```python
fact(0)
```

```python
fact(3)
```

```python
fact(4)
```

```python
def palindrome(String):
    if len(String)<=1:
        return "Palindrome";
    else:
        if String[0]==String[-1]:
            return palindrome(String[1:-1]);
        else:
            return "Not Palidrome";
```

```python
palindrome("ab")
```

```python
palindrome('madam')
```

```python
palindrome('noon')
```

```python
def fib(m):
    if m==0 or m==1:
        return 1;
    else:
        return fib(m-1)+fib(m-2);
```

```python
fib(12)
```

#### Fibonacci Series

```python
def fib(a=0,b=1):
    if b<20:
        print(a,end=' ');
        a,b=b,a+b;
        fib(a,b);
```

```python
fib();
```
