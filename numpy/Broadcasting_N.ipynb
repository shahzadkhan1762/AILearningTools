### vector quantization

```python
from numpy import array,argmin,sqrt;
```

```python
observation=array([111.0,188.0]);
codes=np.array([[102.0,203.0],
                [132.0,193.0],
                [45.0,155.0], 
                [57.0,173.0]]);
```

```python
diff=codes-observation;
```

```python
diff.shape
```

```python
dist=sqrt(sum(diff**2,axis=1));
```

```python
dist.shape
```

```python
argmin(dist)
```

```python
dist[argmin(dist)]
```
