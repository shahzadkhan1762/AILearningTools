#### Prime number function

```python
def prime_Number(num):
    prime=False;
    if num<2:
        return False;
    elif num==2:
        return True;
    else:
        for i in range(2,num):
            if num%i==0:
                return False;
        return True;
```

```python
print(prime_Number(11));
```
