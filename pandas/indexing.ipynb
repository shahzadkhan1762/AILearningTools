### Pandas iloc and loc 

```python
import pandas as pd;
import numpy as np;
```

```python
mydict = [{'a': 1, 'b': 2, 'c': 3, 'd': 4},
          {'a': 100, 'b': 200, 'c': 300, 'd': 400},
          {'a': 1000, 'b': 2000, 'c': 3000, 'd': 4000}]
df =  pd.DataFrame(mydict);
```

```python
df
```

```python
df.iloc[0]
```

```python
df.iloc[:, 0]
```

```python
df.iloc[0,0]
```

```python
df.iloc[[0,1]]
```

```python
df.iloc[[0],[1,2]]
```

```python
df.iloc[0:2]
```

```python
df.iloc[1:3, 0:3]
```

```python
df.iloc[slice(0,2)]    ## slice(start , stop , step)
```

```python
df.iloc[slice(0,2), slice(0,2)]
```

```python
df.iloc[[True, True, False]]
```

```python
df.iloc[[True, True, False], [True, False, True, False]]
```

```python
df.iloc[lambda x: x.index%2==0, :]
```

```python
df.iloc[(lambda x: x.index==0), slice(0,3)]
```

### loc

```python
df = pd.DataFrame([[1, 2], [4, 5], [7, 8]],
                  index=['cobra', 'viper', 'sidewinder'],
                  columns=['max_speed', 'shield'])
```

```python
df
```

```python
df.loc['cobra']
```

```python
df.loc['cobra':]
```

```python
df.loc[['cobra','sidewinder'], ['max_speed']]
```

```python
df.iloc[lambda x: x.index=='cobra']
```

```python
df.loc[df['max_speed']<2]
```

```python
df
```

```python
df.loc[df['shield']>6]
```

```python
df.loc[df['shield'] == 8, 'shield'] = 0
```

```python
df
```

```python
list(df.items())
```

```python
for label, content in df.items():
    print('Label', label);
    print('Content','\n', content);
```
