### Type Conversion

- Type conversion is the process of converting a data type into another data type.

#### Implicit Type Conversion

* Implicit type conversion is performed by a Python interpreter only.

Example

```python
num1=43;
num2=65.3;
result=num1+num2;        # value num will implicitly covert to float
print(result);
print(type(result));
```

#### Expicit Type Conversion

* Explicit type conversion is performed by the user by explicitly using type conversion functions in the program code.

Example

```python
num1=int(input("Enter num1 : "));             # int() function convert input(string) to int
num2=int(input("Enter num2 : "));
result=num1+num2;
print("The sum is : "+str(result));           # str() Fucntion convert result(int) to string format
```
