```python
import random
def guessing_game():
    attempts=1;
    guess_number=int(input("Guess a number from 1 to 100 : "));
    rand_number=random.randint(1,100);
    while guess_number!=rand_number:
        attempts+=1;
        if rand_number<guess_number:
            guess_number=int(input("Guess Lower : "));
        else:
            guess_number=int(input("Guess Higher : "));
            
    if rand_number==guess_number:
            print('Correct! in ',attempts,"attempts");
```

```python
guessing_game();
```
