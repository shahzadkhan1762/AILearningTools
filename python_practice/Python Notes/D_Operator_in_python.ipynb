### Operators

Operators are used to perform operations on variables and values. Python has the following operators.

- Arithmetic Operators
- Comparison Operators
- Logical Operators
- Bitwise Operators
- Assignment Operators
- Identity Operators
- Membership Operators

#### Arithmetic Operators

```python
a=54;
b=34;
```

```python
a+b
```

```python
a-b
```

```python
a*b
```

```python
a/b
```

```python
a%b
```

```python
a**b
```

```python
a//b
```

```python
a!=b
```

#### Identity Operators

```python
a=5;
b=5;

print(a is b);
```

```python
a='hello';
b='hello';
print(a is b)
```

```python
a=[1,2,3];
b=[1,2,3];
print(a is b)
```

```python
a='Hello-world';
b='Hello-world';
print(a is b);
```

```python
a is not b
```
