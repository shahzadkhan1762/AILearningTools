### Connection

```python
# con=oracle.connect("username/password@localhost");
```

```python
con=oracle.connect("scott/Abbas_2004@localhost:1521/ORCL");
```

### Cursor

**cursor(): To execute a SQL query and to provide results some special object is required that is nothing but cursor() object.**

```python
cursor=con.cursor();
```

### execute/executemany method:

```python
# cursor.execute("SQL Query")        --->   To execute single query
# cursor.executemany("SQL Query")    --->   to execute a single query with multiple bind variables/place holders.
```

```python
cursor.execute("select * from dept");
```

### fetchone(), fetchmany(int), fetchall():

**fetchone() : This method is used to fetch one single row from the top of the result set.**

**fetchmany(int) : This method is used to fetch a limited number of rows based on the argument passed in it.**

**fetchall() : This method is used to fetch all rows from the result set.**

### Example:

**The Query is executed again and again because when we fetch a row from cursor the row is deleted from the cursor.**

```python
cursor.execute('select * from dept');
```

```python
print(cursor.fetchone());
```

```python
cursor.execute('select * from dept');
```

```python
print(cursor.fetchmany(2));
```

```python
cursor.execute('select * from dept');
```

```python
rows=cursor.fetchall();
for row in rows:
    print(row);
```

```python
cursor.execute('select * from dept');
```

```python
for row in cursor.fetchmany(2):
    print(row);
```
