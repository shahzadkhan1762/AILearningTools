#### Arrays are fast

```python
import time
```

```python
a=range(100000);
b=range(100000,200000);

start_time=time.time();

c=[(x,y) for x,y in zip(a,b)];

end_time=time.time();
```

```python
print(end_time-start_time);
```

```python
a=np.arange(100000);
b=np.arange(100000,200000);

start_time=time.time();

c=a+b;

end_time=time.time();
```

```python
print(end_time-start_time);
```
