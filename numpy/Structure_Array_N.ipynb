### Structure Array

- Structured arrays are ndarrays whose datatype is a composition of simpler datatypes organized as a sequence of named fields.

```python
import numpy as np;
```

```python
arr1=np.array([('kamran',4.5,50),('usman',5,60),('wahid',5.4,61)], dtype=[('name','|U10'),('hight','f8'),('wight','i4')]);
```
