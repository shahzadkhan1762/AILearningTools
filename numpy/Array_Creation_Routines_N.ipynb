## Array Creation Routines

### Ones and zeros

```python
import numpy as np;
```

#### numpy.empty(shape, dtype=float, order=’C’)

```python
empty = np.empty((3,4), dtype=np.half);
empty
```

```python
empty=np.empty((3,4,5), dtype='b');     # byte integer data type
empty
```

#### numpy.empty_like(a, dtype=None, order=’K’, subok=True)
- Return a new array with the same shape and type as a given array.
- The subok parameter in the numpy.empty_like() function determines whether the returned array should be of a subclass of the input array's class.

```python
empty_like=np.empty_like(empty);
empty_like
```

```python
empty_like=np.empty_like((3,3));
print(empty_like);
print(empty_like.dtype);
print(empty_like.size);
```

```python
empty_like=np.empty_like(empty, dtype='>h', subok=True);   # subok is true by default 
print(empty_like);
print(empty_like.dtype);
print(empty_like.dtype.byteorder);
print(type(empty_like));
```

```python
empty_like=np.empty_like(empty,subok=False);
print(empty_like);
print(type(empty_like));
```

```python
import numpy as np;

class SubArray(np.ndarray):
    pass

arr=SubArray([6,3,7]);

empty_like=np.empty_like(arr, subok=False);
print(type(empty_like));

empty_like=np.empty_like(arr, subok=True);
print(type(empty_like));
```

#### numpy.eye(N, M=None, k=0, dtype=<type ‘float’>)
- Return a 2-D array with ones on the diagonal and zeros elsewhere.
- N : Number of rows in the output
- M : int,optional  Number of columns in the output if None, defaults to N.
- K : int,optional  Index of the diagonal: 0 (the default) refers to the main diagonal, a positive value refers
to an upper diagonal, and a negative value to a lower diagonal.

```python
eye=np.eye(4);
print(eye);
print(eye.dtype);
```

```python
eye=np.eye(3,4);
print(eye);
print(eye.dtype);
print(eye.dtype.byteorder);
```

```python
eye=np.eye(4,4,0,'H');
print(eye);
print(eye.dtype);
```

```python
eye=np.eye(3,5,1,'l');
print(eye);
print(eye.dtype);
```

```python
eye=np.eye(4,4,-1,'f');
print(eye);
print(eye.dtype);
```

```python
eye=np.eye(4,4,-3,'f');
print(eye);
print(eye.dtype);
```

```python
eye=np.eye(4,4,-5,'f');
print(eye);
print(eye.dtype);
```

```python
eye=np.eye(N=5,M=3,k=2,dtype=np.int64);
print(eye);
print(eye.dtype);
```

#### numpy.identity(n, dtype=None)
- Return the identity array

```python
i=np.identity(4);
print(i);
print(i.dtype);
```

```python
i=np.identity(5,'g');   # longdouble float data type
print(i);
print(i.dtype);
```

```python
empty_like=np.array(i);
print(empty_like);
print(empty_like.dtype);
```

#### numpy.ones(shape, dtype=None, order=’C’)
- Return a new array of given shape and type, filled with ones.

```python
ones=np.ones(10,dtype=np.float32);
print(ones);
print(ones.dtype);
```

```python
ones=np.ones((4,5,3), dtype='q');
print(ones);
print(ones.dtype);
print(ones.dtype.byteorder);
```

```python
ones=np.ones((3,3), dtype=np.ubyte)*4;
print(ones);
print(ones.dtype);
```

#### numpy.ones_like(a, dtype=None, order=’K’, subok=True)
- Return an array of ones with the same shape and type as a given array

```python
arr=np.empty((4,3), 'd');
ones_like=np.ones_like(arr);
print(ones_like);
print(ones_like.dtype);
```

```python
arr=np.arange(1,21,1,'i').reshape(2,5,2);
print(arr);
ones_like=np.ones_like(arr);
print(ones_like);
print(ones_like.dtype);
```

#### numpy.zeros(shape, dtype=float, order=’C’)

```python
zeros=np.zeros(5,dtype=np.single);
print(zeros);
print(zeros.dtype);
```

```python
zeros=np.zeros((4,3), dtype='>u2');     # unsigned int 2byte   (np.ushort)
print(zeros);
print(zeros.dtype);
```

#### numpy.zeros_like(a, dtype=None, order=’K’, subok=True)
- numpy.zeros_like(a, dtype=None, order=’K’, subok=True)

```python
arr=np.eye(4,3,1,dtype='f2')
print(arr);
print();
z_like=np.zeros_like(arr);
print(arr);
print(arr.dtype);
```

```python
arr=np.linspace(3,10,8);
print(arr);
print();
z_like=np.zeros_like(arr,dtype=np.longlong,subok=False);
print(z_like);
print(z_like.dtype);
```

```python
arr=np.array([4,5,2,6,2,5]);
print(arr);
print();
z_like=np.zeros_like(arr, dtype=np.ushort);
print(z_like);
print(z_like.dtype);
```

```python
arr=np.ones((2,2), dtype='B');
print(arr);
print(arr.size)
print();
z_like=np.zeros_like(arr, dtype=[('Age','b'),('Marks','f2')]);
z_like
```

#### numpy.full(shape, fill_value, dtype=None, order=’C’)
- Return a new array of given shape and type, filled with fill_value.

```python
full=np.full((3,4),100,dtype='f');
print(full);
print(full.dtype);
```

```python
full=np.full((2,5), 340/550*100, 'd');
print(full);
print(full.dtype);
```

#### numpy.full_like(a, fill_value, dtype=None, order=’K’, subok=True)
- Return a full array with the same shape and type as a given array.

```python
arr=np.arange(24).reshape(3,4,2);
full_like=np.full_like(arr,np.nan);
print(full_like);
print(full_like.dtype)
```

```python
full_like=np.full_like(arr, 200, dtype=np.short, subok=False);
print(full_like);
print(full_like.dtype);
```

### Numarical Ranges

```python
import numpy as np;
```

#### numpy.arange([start], stop[, step ], dtype=None)


```python
arange=np.arange(1,31,1,dtype='i');
print(arange);
print(arange.dtype);
```

```python
arange=np.arange(stop=15);
print(arange);
print(arange.dtype);
```

```python
arange=np.arange(stop=15,step=1,dtype=np.int_).reshape(3,5);
print(arange);
print(arange.dtype);
```

#### numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)
- endpoint : bool, optional
    - If True, stop is the last sample. Otherwise, it is not included. Default is True.
- retstep : bool, optional
    - If True, return (samples, step), where step is the spacing between samples.

```python
l=np.linspace(1,10,10,False,True,dtype='f');
l
```

```python
l=np.linspace(3,20,24);
print(l);
print(l.dtype);
```

```python
l=np.linspace(3,150,24,endpoint=True,retstep=False, dtype='d').reshape(2,12);
print(l);
print(l.dtype);
```

```python
import matplotlib.pyplot as plt;

N=8;
y=np.zeros(8);
x1=np.linspace(0,10,N, endpoint=True);
plt.plot(x1,y,'o');
```

```python
x2=np.linspace(0,10,N,endpoint=False);
plt.plot(x2,y+0.02,'*');
```

```python
x1=np.linspace(0,20,10,endpoint=True);
y1=np.linspace(0,20,10,endpoint=True);
plt.plot(x1,y1,'o');
```

```python
x1=np.linspace(1,15,10);
x2=np.linspace(-1,13,10);
y=np.linspace(0,10,10);
plt.plot(x1,y,'*');
plt.plot(x2,y,'*');
plt.xlim([-5,15])
```

#### numpy.logspace(start, stop, num=50, endpoint=True, base=10.0, dtype=None)

```python
log=np.logspace(0,9,10);
print(log);
print(log.dtype);
```

```python
lin=np.linspace(1,10,10);
print(lin);
log=np.logspace(1,10,10);
print(log);
```

```python
# space between the elements of logspace array

print("(10**2) - (10**1)  : ", (1.e+2)-(1.e+1));
print("(10**3) - (10**2)  : ", (1.e+3)-(1.e+2));
print("(10**4) - (10**3)  : ", (1.e+4)-(1.e+3));
print("(10**5) - (10**4)  : ", (1.e+5)-(1.e+4));
print("(10**6) - (10**5)  : ", (1.e+6)-(1.e+5));
print("(10**7) - (10**6)  : ", (1.e+7)-(1.e+6));
print("(10**8) - (10**7)  : ", (1.e+8)-(1.e+7));
print("(10**9) - (10**8)  : ", (1.e+9)-(1.e+8));
print("(10**10) - (10**9) : ", (1.e+10)-(1.e+9));
```

```python
log=np.logspace(1,10,10,endpoint=True,base=2.0);
print(log)
```

```python
x1=np.logspace(1,3,10, endpoint=True);
x2=np.logspace(1,3,10, endpoint=True);
y=np.logspace(1,3,10);
plt.plot(x1,y,'o');
plt.plot(x2,y+100,'o');
```
