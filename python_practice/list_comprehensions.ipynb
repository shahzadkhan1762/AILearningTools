### list comprehensions

```python
square=[x*2 for x in range(10)]
```

```python
square
```

```python
combs=[(x,y) for x in [1,2,3] for y in [3,4,5] if x!=y]
```

```python
combs
```

```python
type(combs[0])
```

```python
t=0;
for i in combs:
    for j in combs[t]:
        print(j,end=' ');
    t+=1;
    print();
```

```python
numbers=[3,-4,0,-2,6,-9];
```

```python
[x for x in numbers]
```

```python
[abs(x) for x in numbers]
```

```python
[x for x in numbers if x>0]
```

```python
[(x,x*2) for x in numbers]
```

```python
numbers=[[2,5,6],[8,3,2],[6,2,7]]
```

```python
[j for i in numbers for j in i]
```
