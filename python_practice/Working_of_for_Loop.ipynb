## Understanding How For Loop Works

- Iterable is an object which can be looped over or iterated over with the help of a for loop.
- Each Iterable object has an iterator.
- Each Iterator has a next function. which is used to retrieve the next element from an iterator.
- When we use a for loop on an iterable it simply first look for iterator of the iterable object used in for loop.
- And then call next the next function using iterator.

### Making For Loop

```python
def for_loop(iterable):
    obj_iter=iter(iterable)
    while True:
        try:
            print(next(obj_iter));
        except StopIteration:
            break;
```

```python
for_loop([1,2,3,4])
```

```python
for_loop((3,4,56))
```
