### Break Statement

```python
student=['abbas','qasim','khalid','waseem','kamran','sohail'];
for i in range(len(student)):
    if student[i]=='waseem':
        print('Available');
        break;
```

### Continue Statement

```python
for i in range(len(student)):
    if student[i]=='waseem':
        continue;
    print(student[i]);
```
