## Tuple

- Tuple is an order,immutable collection of elements. It is similar to a list in that can store multiple elements, but unlike list tuple can't be modified after it is created.

### Creating Tuple

```python
T=(12,33,45,67);
```

```python
type(T)
```

```python
T1=()
```

```python
type(T1)
```

```python
# creating a tuple with only one element will create only a single variable.
T2=(23);
type(T2)
```

```python
# solution
T2=(23,)
```

```python
type(T2)
```

```python
T4=tuple([2,3,4,5,6]);
```

```python
T4
```

```python
type(T4)
```

```python
T5=('khan',23,'abbas',20,'usman',20);
```

```python
T5
```
