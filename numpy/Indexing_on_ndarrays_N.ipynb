### Striding

- Striding is a mechanism by which Numpy accessing and operating on arrays. Stides define the number of step in each dimension when iterating over the array.

```python
import numpy as np;
```

```python
arr_strides=np.array([[[23,45,54],[64,23,64]],[[56,23,74],[54,94,45]],[[95,58,44],[95,86,38]]]);
```

```python
arr_strides
```

```python
arr_strides.strides          # it means that to move from one elements to another in the dimension(1) numpy jumps total 24Bytes, and to move from one element to another in dimension(2,row), Numpy jumps total 12Bytes. and to move from one element to the next in the second dimension Numpy Jumps total 4bytes.  
```

```python
# Another Example
```

```python
arr_strides.shape=(3,6);
```

```python
arr_strides
```

```python
arr_strides.strides
```
