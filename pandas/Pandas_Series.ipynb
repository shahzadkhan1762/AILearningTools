### Pandas Series

```python
import numpy as np;
import pandas as pd;
```

```python
D = {'a':234,'b':532,'c':345};
Series = pd.Series(D, index=['a','b','c']);
Series
```

```python
Series = pd.Series([23,542,23,53], index=[1,2,3,4]);
Series
```

```python
D = {'A':433, 'B':633, 'C':432};
Series = pd.Series(D, index=[1,4,5]);
Series
```

```python
D = [34,53,23,53,24];
Series = pd.Series(D, copy=False);
Series.iloc[0]=100;
print("Series : \n",Series);
print("D : ",D);
```

```python
D = [532,744,723,635,156];
Series = pd.Series(D, copy=True);
Series.iloc[2]=100;
print(Series);
print(D);
```

```python
D = np.array([4,34,23,54,23,53]);
Series = pd.Series(D, copy=False);
Series.loc[0] = 100;
print(" D : ",D);
print(" Series : \n",Series);
```
