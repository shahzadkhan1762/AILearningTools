 Class to store the detail of user 

```python
class User():
    
    def __init__(self, first_name='N/A', last_name='N/A', age='N/A', address='N/A', password='N/A'):
        self.first_name=first_name;
        self.last_name=last_name;
        self.age=age;
        self.address=address;
        self.password=password;
        
    def desc_user(self):
        print("Detail of User");
        print("First Name : "+self.first_name);
        print("Last Name : "+self.last_name);
        print("Age : ",self.age);
        print("Address : "+self.address);
        print("Password : "+self.password);
        
    def great_user(self):
        print("welcome "+self.first_name+" "+self.last_name+"!");
```

```python
user=User('Muhammad',"Abbas",20,"Hangu","sjdfkljsd");
```

```python
print(user.first_name)
```

```python
print(user.last_name)
```

```python
print(user.age)
```

```python
print(user.address)
```

```python
print(user.password)
```

```python
user.desc_user();
```

```python
user.great_user();
```
