## Function in python

### Factorial of a Number 

```python
def fact(num):
    fac=1;
    if num==0:
        return 1;
    for i in range(1,num+1):
        fac *=i;
    
    return fac;
```

```python
print(fact(4));
```

```python
print(fact(0));
```

```python
print(fact(1));
```

```python
print(fact(10));
```

### Factorial using recursion

```python
def factorial(num):
    if num==0:
        return 1;
    return num*factorial(num-1);
```

```python
print(factorial(4));
```

```python
num=int(input("Enter a number to find its factorial : "));
result=factorial(num);
print("The facotrial of {} is {} ".format(num,result))
```

### Count even and odd numbers in a lis 

```python
def count(lst):
    even=0;
    odd=0;
    for i in lst:
        if i%2==0:
            even+=1;
        else:
            odd+=1;
            
    return even,odd;
```

```python
lst=[4,63,6,2,7,23,74,77,23,93,23,5,8,23,8,23]
even,odd=count(lst);
print("Evne Number are {} & Odd Number are {}".format(even,odd))
```
