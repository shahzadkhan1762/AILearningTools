```python
class Restaurant():
    """ class to  represent a Restaurant """
    
    def __init__(self,restaurant_name, cuisine_type):
        self.restaurant_name=restaurant_name;
        self.cuisine_type=cuisine_type;
        self.served_customer=0;
        
    def describe_res(self):
        print("The name of restaurant is : "+self.restaurant_name);
        print("Restaurant type is : "+self.cuisine_type);
        print("Number of Customer "+self.restaurant_name+" served : ",self.served_customer);
        
    def set_number_served(self,number):
        self.served_customer=number;
    
    def incr_number_served(this, number):
        this.served_customer+=number;
        
    def open_res(self):
        print(self.restaurant_name+" is open.");
```

```python
restaurant=Restaurant("islamia restaurant","full service");
```

```python
restaurant.describe_res();
```

```python
restaurant.served_customer=30;
```

```python
print(restaurant.served_customer);
```

```python
restaurant.set_number_served(500);
```

```python
restaurant.describe_res();
```

```python
restaurant.incr_number_served(23);
```

```python
print(restaurant.served_customer);
```

### Creating Child Class of Restaurant

```python
class IceCreamStand(Restaurant):
    
    def __init__(self, restaurant_name, cuisine_type):
        super().__init__(restaurant_name, cuisine_type);
        self.flavors=['vanilla','chocolate','coffee','chocolate chip','stawberry'];
        
    def print_flavors(self):
        print("List of ice cream flavors");
        print(self.flavors);
        
```

```python
icecreamstand=IceCreamStand('Peshawar Ice Cream','Ice Cream Stand');
```

```python
icecreamstand.print_flavors();
```
