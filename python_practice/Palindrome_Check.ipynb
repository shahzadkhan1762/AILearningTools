```python
def is_palindrome(string):
    clean_str='';
    for i in string:
        if i.isalnum():
             clean_str+=''.join(i.lower());
    return clean_str==clean_str[::-1];
```

```python
is_palindrome('hi')
```

```python
is_palindrome('radar')
```

```python
is_palindrome('refer')
```
