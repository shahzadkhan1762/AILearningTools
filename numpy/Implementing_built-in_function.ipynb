### My Own Find Method

```python
def Find(s,ch,n=1):
    """
    Find the index of character 
    take three argumet
      * s mean a string
      * ch mean a character
      * number of occurance
    """
    t=1;
    for c in range(len(s)):
        if ch==s[c]:
            if n==t:
                return c;
            else:
                t+=1;
    return -1;
```
