## Universal Functions
- Universal functions, often abbreviated as "ufuncs," are a key feature of the NumPy library in Python. Ufuncs provide a way to perform element-wise operations on arrays, broadcasting the operations across the entire array without the need for explicit looping.

```python
import numpy as np;
```

#### np.add(x1,x2[, out])
- Add arguments element-wise.

```python
np.add(43.2,724.5)
```

```python
np.add(34/2+4, 15*2)
```

```python
arr1=np.arange(12);
arr2=np.arange(12);
np.add(arr1,arr2)
```

```python
arr1=np.linspace(1,2,10, endpoint=True);
arr2=np.logspace(1,4,10, endpoint=True);
print(np.round(arr1,2), end='\n\n');
print(np.round(arr2,2), '\n\n');
print(np.round(np.add(arr1,arr2), 2));
```

```python
arr1=np.ones((3,5), dtype='f4');
arr2=np.ones_like(arr1);
np.add(arr1,arr2)
```

```python
arr1=np.zeros((3,6), dtype='e');
print(arr1,'\n\n')
arr2=np.ones((3,1), dtype='f');
print(arr2,'\n\n')
np.add(arr1,arr2)
```

#### np.reciprocal(x[, out])
- Return the reciprocal of the argument element-wise.
- Calculates  1/x;
- Note: This function is not designed to work with integers.

```python
np.reciprocal(1.0)
```

```python
np.reciprocal(2.0)
```

```python
arr1=np.array([35,73,23,72,23,63], dtype='>e');
np.reciprocal(arr1)
```

```python
arr1=np.arange(1,20,1, 'd');
np.round(np.reciprocal(arr1), 4)
```

```python
arr1=np.arange(11,23,1, dtype='f').reshape(2,6);
np.reciprocal(arr1)
```

#### np.divide(x1, x2[, out])
- multiply arguments element-wise

```python
np.divide(43,63)
```

```python
np.divide(12+3, 8-5)
```

```python
arr1=np.arange(1,21,1, dtype='h').reshape(4,5);
arr2=np.copy(arr1);
np.divide(arr1,arr2)
```

#### np.power(x1, x2[, out])

#### np.subtract(x1, x2[, out])

#### np.true_divide(x1, x2[, out])
- Returns a true division of the inputs, element-wise

```python
np.true_divide(23,3)
```

```python
np.true_divide(2352,72)
```

```python
arr1=np.arange(12);
np.true_divide(arr1, 3)
```

```python
arr1=np.ones((3,3), dtype='f')*5;
arr2=np.ones_like(arr1, dtype='f')*2;
np.true_divide(arr1,arr2)
```

#### np.floor_divide(x1, x2[, out])
- Return the largest integer smaller or equal to the division of the inputs.

```python
np.floor_divide(10,4)
```

```python
np.floor_divide(334,53)
```

```python
arr1=np.arange(1,12);
arr2=np.arange(1,12)*3;
np.floor_divide(arr1,arr2)
```

```python
arr1=np.linspace(1,21,20).reshape(4,5);
arr2=np.linspace(1,21,4).reshape(4,1);
np.floor_divide(arr1,arr2)
```
