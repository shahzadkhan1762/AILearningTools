```python
# ndarray.conj()
# Complex-conjugate all elements.

c_arr = np.array([23+5j, -3-5j, 23+1j, 4-5j]);
print(c_arr);
print(c_arr.dtype);
print(c_arr.size);

# c_arr.con()   change the sign of complex imaginary part
c_arr.conj()
```

```python
c_arr = np.arange(10, dtype=np.complex_)
print(c_arr);
print(c_arr.dtype);
print(c_arr.size);

# c_arr.con()   change the sign of complex imaginary part
c_arr.conj()
```

```python
# ndarray.cumprod(axis=None, dtype=None, out=None)
# Return the cumulative product of the elements along the given axis.

arr = np.array([[2,4,2],[5,2,6],[6,4,1]]);
print(arr);
print();
print(arr.cumprod())
```

```python
print(arr);
print();
print(arr.cumprod(axis=0, dtype=np.int32));
```

```python
print(arr);
print();
print(arr.cumprod(axis=1, dtype=np.float16))
```

```python
# ndarray.cumsum(axis=None, dtype=None, out=None)
# Return the cumulative sum of the elements along the given axis.

arr = np.arange(6).reshape(2,3);
print(arr);
print();
print(arr.cumsum());
print();
print(arr.cumsum(axis=0, dtype='e'));
print();
print(arr.cumsum(axis=1, dtype='f'));
```

```python
arr = np.linspace(1, 3, 20, dtype=np.half).reshape(2,2,5);
print(arr);
print(end='\n\n');
print(arr.cumsum(axis=-1));
```

```python
# ndarray.choose(choices, out=None, mode=’raise’)
# Use an index array to construct a new array from a set of choices.
# the size of choices must be matched to the index array
```

```python
choices = [[0, 1, 2, 3], [10, 11, 12, 13],[20, 21, 22, 23], [30, 31, 32, 33]];
np.choose([2, 3, 1, 0], choices)
```

```python
index_arr = np.array([3,2,0,1]);
choice = ([[34,63,23,52], [23,632,235,23], [53,23,63,23], [43,63,23,53]])

index_arr.choose(choice)
```

```python
index_arr = np.array([1,2,0]);
choices = (np.array([95,32,63]), np.array([96,44,100]), np.array([77,29,40]));

index_arr.choose(choices)
```
