#### zip();

**In Python, the zip() function is a built-in function that takes multiple iterable objects as input and returns an iterator of tuples, where the first item in each passed iterable is paired together, the second item in each passed iterable is paired together, and so on.**

```python
## zip(iterable,iterable,.....)
```
