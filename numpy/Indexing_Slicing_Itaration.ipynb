#### Slicing

```python
arr3=np.array([2,3,5,6,73,7]);
```

```python
print(arr3[:])
```

```python
print(arr3)
```

```python
print(arr3[2:])
```

```python
print(arr3[2:3])
```

```python
print(arr3[::-1])
```

```python
print(arr3[-4:])
```

```python
print(arr3[-1])
```

```python
for i in arr3[::-1]:
    print(i,end=' ')
```
