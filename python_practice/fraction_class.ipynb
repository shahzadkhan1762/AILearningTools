### How to built our own data type

#### Here we will create a data type called fraction

```python
class Fraction:
    
    def __init__(self,num,den):
        self.num=num;
        self.den=den;
        
    def __str__(self):
        return "{}/{}".format(self.num,self.den);
    
    def __add__(self,other):
        tnum=self.num*other.den+self.den*other.num; 
        tden=self.den*other.den;
        tfraction=Fraction(tnum,tden);
        #return "{}/{}".format(tnum,tden);
        return tfraction;
    
    def __sub__(self,other):
        tnum=self.num*other.den-self.den*other.num;
        tden=self.den*other.den;
        tfraction=Fraction(tnum,tden);
        #return "{}/{}".format(tnum,tden);
        return tfraction;
    
    def __mul__(self,other):
        tnum=self.num*other.num;
        tden=self.den*other.den;
        tfraction=Fraction(tnum,tden);
        #return "{}/{}".format(tnum,tden);
        return tfraction;
    
    def __truediv__(self,other):
        tnum=self.num*other.den;
        tden=self.den*other.num;
        tfraction=Fraction(tnum,tden);
        #return "{}/{}".format(tnum,tden);
        return tfraction;
```

```python
f=Fraction(2,3);
```
