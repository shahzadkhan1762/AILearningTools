#### if Statement

```python
x=int(input("please enter an integer : "));
if x<0:
    x=0;
    print("Negative changed to 0");
elif x==0:
    print("Zero");
elif x==1:
    print("One");
else:
    print("more");
```

### enumerate Function

**When looping through a sequence, the position index and corresponding value can be retrieved at the same time using the enumerate() function.**

```python
Tuple=("C++","Python","Java","R","JavaScript","Cabul","CSS");
for index,value in enumerate(Tuple):
    print("Index : ",index,"    ","Value : ",value);
```

**Reverse the Tuple**

```python
for index,value in enumerate(Tuple[::-1]):
    print("Indes : ",index,"   ","Value : ",value);
```

### else statement with for loop    (break Statement)

 **loopâ€™s else clause runs when no break occurs**

```python
num=int(input("Enter a number : "));
for n in range(2,int(num/2+1)):
    if num%n==0:
        print("Not prime");
        break;
else:
    print("prime number");
```

#### Continue Statement

```python
for num in range(1,15):
    if num%2==0:
        print("Event number found : ",num);
        continue;
    print("Odd Number found : ",num)
```
