### iter() Function

```python
L=[34,53,36,23,85,24,75];
```

```python
try:
    next(L);
except TypeError as ex:
    print("Type Error : ",ex);
```

```python
iter_obj=iter(L);
```

```python
iter_obj
```

```python
for i in L:                 
    print(next(iter_obj))
```

### List compreshensions

```python
L
```

```python
square=[];
for num in L:
    sq=num**2;
    square.append(sq);
square
```

```python
square=[sq**2 for sq in L]
square
```

```python
square=[];
for num in L:
    sq=num**2;
    if sq%2==0:
        square.append(sq);

square
```

```python
square=[sq**2 for sq in L if (sq**2)%2==0]
square
```

#### walrus operator

```python
square=[sq for num in L if (sq:= num**2)%2==0]
square
```

```python
not (3**2)%2
```

```python
not (4**2)%2
```

```python
square=[sq for num in L if not(sq:= num**2)%2]
square
```

### Dictionary Comprehension

```python
d={char:ord(char) for char in 'abcdefghi'}
d
```

```python
# Set
s={ord(char) for char in 'abcdefgh'};
s
```

```python
# generator expression
g=(el**2 for el in range(5));
g
```

```python
for i in g:
    print(i)
```

```python
# Tuple
t=tuple(ord(char) for char in 'abcdefgh');
t
```
