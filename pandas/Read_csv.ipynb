### The converters Parameter
- dict of {HashableCallable}, optional
- Keys can either be column labels or column indices.

```python
df = pd.read_csv('csv_like_file.txt', sep='\t', converters={'Fare': lambda x: float(x)*10});
df
```

```python
fare_ctr = lambda x: float(x)*10;
age_ctr = lambda x: int(x)+10;

df = pd.read_csv('csv_like_file.txt', sep='\t', converters={'Age':age_ctr, 'Fare':fare_ctr});
df
```

```python
fare_ctr = lambda x: float(x)*10;
age_ctr = lambda x: int(x)+10;

ctrs = {'Age':age_ctr, 'Fare':fare_ctr};

df = pd.read_csv('csv_like_file.txt', sep='\t', converters=ctrs);
df
```

```python
fare_ctr = lambda x: float(x)*10;
age_ctr = lambda x: int(x)+10;

ctrs = {1:age_ctr, 2:fare_ctr};

df = pd.read_csv('csv_like_file.txt', sep='\t', converters=ctrs);
df
```

```python
# using converters in datetime 
df = pd.read_csv("Dates_time.csv", parse_dates=[0]);
df
```

### keep_default_na 

- bool, default True

- If keep_default_na is True, and na_values are specified, na_values is appended to the default NaN values used for parsing.

- If keep_default_na is True, and na_values are not specified, only the default NaN values are used for parsing.

- If keep_default_na is False, and na_values are specified, only the NaN values specified na_values are used for parsing.

- If keep_default_na is False, and na_values are not specified, no strings will be parsed as NaN.

- Note that if na_filter is passed in as False, the keep_default_na and na_values parameters will be ignored.

```python
pd.read_csv('Employee.csv', keep_default_na=True, na_values=[',','No'])[0:10]
```

```python
pd.read_csv('Employee.csv', keep_default_na=True)[0:10]
```

```python
pd.read_csv('Employee.csv', keep_default_na=False, na_values=[',','No','Null'])[0:10]
```

```python
pd.read_csv('Employee.csv', keep_default_na=False)[0:10]
```

### na_filter
- used to control whether to detect and filter out missing values (NA/NaN) during the reading of a CSV file.
- Note that if na_filter is passed in as False, the keep_default_na and na_values parameters will be ignored.

```python
pd.read_csv("Employee.csv", na_filter=True)[0:10]
```

```python
pd.read_csv('Employee.csv', na_filter=False)[0:10]           # No changes Origional file
```

```python
pd.read_csv('Employee.csv', na_filter=False, keep_default_na=True, na_values=[',',' '])[0:10]  # No change Origional file
```
