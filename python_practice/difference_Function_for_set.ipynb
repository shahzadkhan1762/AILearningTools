```python
def difference(S1,S2):
    set_list=list(S1);
    deff=[];
    for i in set_list:
        if i not in S2:
            deff.append(i);
    return set(deff);
```

```python
S1={3,5,7,9,11};
S2={2,4,6,8,10}
difference(S1,S2)
```

```python
S1={2,3,5,7,8};
S2={1,3,4,7,10};
difference(S1,S2)
```
