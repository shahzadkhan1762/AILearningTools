```python
np.datetime64('2023')==np.datetime64('2023-01-04')
```

```python
np.datetime64('2005') == np.datetime64('2005-01-01')
```

### Datetime and timedelta Arithmetic

```python
np.datetime64('2023-08-31')-np.datetime64('2023-08-01')
```

```python
np.datetime64('2023')-np.datetime64('2000')
```

```python
np.datetime64('2023-08-31')-np.datetime64('2023')
```

```python
np.datetime64('2023')+np.timedelta64(32,'D')
```

```python
np.datetime64('2023-08-31')-np.timedelta64(31,'D')
```

```python
np.datetime64('2012-08-31')+np.timedelta64(24,'h')
```

```python
np.datetime64('2012')+np.timedelta64(12,'h')
```

```python
np.datetime64('2023')+np.timedelta64(435,'ms')
```

```python
np.datetime64('2023')+np.timedelta64(12,'M')
```

```python
np.timedelta64(4,'W')/np.timedelta64(1,'D')
```

```python
np.timedelta64(10,'Y')/np.timedelta64(1,'M')
```

```python
a=np.timedelta64(3,'Y')
np.timedelta64(a,'M')
```

###  Business Day Functionality

- np.busday_offset()
    - 
   The numpy.busday_offset function is a part of the NumPy library in Python. It's used to calculate the offset from a given date (start date) by a certain number of business days.
   - For example, if you have a starting date of August 31, 2023, and you apply an offset of 5 using the numpy.busday_offset function, it means you are calculating a new date that is 5 business days after August 31, 2023.

```python
np.busday_offset('2023-08-31',4)
```

```python
np.busday_offset('2023-08-31',10)
```

```python
np.busday_offset('2023-08-31',0)
```

```python
np.busday_offset('2023-08-31',0,roll='raise')
```

```python
np.busday_offset('2023-08-31',2,roll='raise')
```

```python
np.busday_offset('2023-08-31', 20, roll='backward')
```

```python
np.busday_offset('2023-08-31', 20, roll='forward')
```

```python
np.busday_offset('2023-08-31', 10, roll='forward', weekmask='1111110')
```

```python
np.busday_offset('2023-08-31', 10, roll='forward', weekmask='1111111')
```
