```python
def return_sum(func,L):
    result_sum=0
    for i in L:
        if func(i):
            result_sum+=i;
    return result_sum;
```

```python
L=[23,52,23,74,34,73,34,63,83,34];
print(return_sum((lambda x: x%2==0),L));
print(return_sum((lambda x: x%2!=0),L));
print(return_sum((lambda x: x%3==0),L));

```
