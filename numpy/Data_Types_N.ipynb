## Data Types


**A data type object (an instance of numpy.dtype class) describes how the bytes in the fixed-size block of memory corresponding to an array item should be interpreted. It describes the following aspects of the data:**

1. Type of the data (integer, float, Python object, etc.)
2. Size of the data (how many bytes is in e.g. the integer)
3. Byte order of the data (little-endian or big-endian)
4. If the data type is structured data type, an aggregate of other data types, (e.g., describing an array item consisting of an integer and a float),
    - what are the names of the â€œfieldsâ€ of the structure, by which they can be accessed,
    - what is the data-type of each field, and
    - which part of the memory block each field takes
    
5. If the data type is a sub-array, what is its shape and data type.

**Numpy Data types are instances of numpy.dtype class.**

### Byte order

- Byteorder refer to the order in which the bytes of a multi-byte value are arranged in memory.

#### Two Byte Order in Numpy
- Little-Endian
    - In little-endian byte order, the least significant byte (LSB) is stored at the lowest memory address, while the most significant byte (MSB) is stored at the highest memory address. This is the byte order used by most Intel and AMD processors.
- Big-Endian
    - In big-endian byte order, the most significant byte (MSB) is stored at the lowest memory address, and the least significant byte (LSB) is stored at the highest memory address. This is used by some other processors and networking protocols.
    
- Example
    - For example, let's consider the 16-bit binary number 0101011011001101. In this binary number, the leftmost 8 bits (01010110) represent the most significant byte (MSB), and the rightmost 8 bits (11001101) represent the least significant byte (LSB). When this number is stored in memory, the byte order depends on the system's endianness.

```python
import numpy as np;
little_endian = np.dtype('<i4');   # '<' represent little-endian byte order
big_endian = np.dtype('>i8');      # '>' represent big-endian byte order

print("Little Endian ",little_endian.byteorder);
print("Big Endian ",big_endian.byteorder);
```
