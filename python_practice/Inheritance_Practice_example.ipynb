```python
class Phone:
    
    def __init__(self,price,brand,camera):
        self.__price=price;                # private
        self.brand=brand;
        self.camera=camera;
    
    def show(self):
        print("Price : ",self.__price);
        print("Brand : ",self.brand);
        print("Camera : ",self.camera)
```

```python
class FeaturePhone(Phone):
    pass;
```

```python
class SmartPhone(Phone):
    
    def show(self):
        print("Price : ",self.__price);
        print("Brand : ",self.brand);
        print("Camera : ",self.camera)
```

```python
smartphone=SmartPhone(3000,'Nokia',10);
```

```python
smartphone.show();
```

```python
# Because the __price is private member of phone class
```

```python
featurephone=FeaturePhone(300000,'IPhone',13);
```

```python
featurephone.show();   ## Because FeaturePhone class has no method show() so show() method of phone class will be callled
```

```python
class Parent:
    
    def __init__(self,num):
        self.__num=num;
        
    def get_num(self):
        return self.__num;
```

```python
class Child(Parent):
    
    def show(self):
        print("This is child class ");
```

```python
son=Child(300);
```

```python
print(son.get_num());
```

```python
son.show();
```

```python
class Parent1:
    
    def __init__(self,num):
        self.num=num;
```

```python
class Child1(Parent1):
    
    def __init__(self,num,var):
        super().__init__(num);
        self.var=var;
    
    def show(self):
        print("Num : ",self.num);
        print("Var : ",self.var);
```

```python
child=Child1(20,300);
```

```python
child.show();
```
