```python
import numpy as np;
```

### numpy.insert(arr, obj, values, axis=None)
- Insert values along the given axis before the given indices.

```python
# help(np.insert);
```

```python
arr = np.arange(10);
arr
```

```python
np.insert(arr, 2, 20)
```

```python
np.insert(arr, 10, 39)
```

```python
arr = np.zeros((3,3), dtype=np.short);
arr
```

```python
np.insert(arr, [1,1], 100)
```

```python
np.insert(arr, 2, 55, axis=1)
```

```python
np.insert(arr, 3, [4,5,6], axis=0)
```

```python
np.insert(arr=arr, obj=(0,2), values=[32,66,34], axis=0)
```

```python
np.insert(arr, 0, 3.3, axis=0)   # type casting
```

```python
np.insert(arr, 2, [False, 343.4, 34])
```

```python
np.insert(arr, 2, [False, 343.4, True], axis=0)
```

### numpy.append(arr, values, axis=None)
- Append values to the end of an array.

```python
arr = np.random.randint(3,100,size=(4,5));
arr
```

```python
np.append(arr, 344)
```

```python
np.append(arr, [[43,52,23,52,23]], axis=0)
```

```python
arr1 = np.array([[3,5],[6,3]]);
arr1
```

```python
np.append(arr1, [[3,4]], axis=0)
```

### numpy.delete(arr, obj, axis=None)
- Return a new array with sub-arrays along an axis deleted. For a one dimensional array, this returns those entries not returned by arr[obj].

```python
arr = np.ones((5,2), dtype='e')
arr
```

```python
np.delete(arr, 1)
```

```python
np.delete(arr, (1,2,3,4,5))
```

```python
arr
```

```python
np.delete(arr, 1, axis=0)
```

```python
np.delete(arr, -1, axis=1)
```
