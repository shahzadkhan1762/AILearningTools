### How to calculate next cpu burst  OS

```python
p_burst=6;
p_guess=10;
alpha=0.5;
next_guess=0;

while True:
    p_burst=int(input("Enter Previous CPU Burst : "));
    next_guess = p_burst*alpha + (1.0-alpha)*p_guess;
    print("Next CPU Burst Guess is : ",next_guess);
    
    p_guess=next_guess;
    
    op=input("Enter E to stop : ");
    if op in 'Ee':
        break;
```
