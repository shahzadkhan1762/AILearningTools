```python
# skip_header and skip_footer argument
```

```python
emp_data=np.genfromtxt(file_name, dtype=data_type, comments='\\', names=True, skip_header=1, skip_footer=5);
emp_data
```

```python
emp_data.size
```

```python
emp_data.shape
```

```python
# using usecols argument
```

```python
emp_data=np.genfromtxt(file_name, dtype=data_type, comments='\\', names=True, usecols=1);
emp_data
```

```python
emp_data=np.genfromtxt(file_name, dtype=data_type, comments='\\', names=True, usecols=[0,1,2,5]);
print(emp_data)
```

```python
emp_data=np.genfromtxt(file_name, dtype=data_type, comments='\\', names=True, usecols=['ENAME','SAL','HIREDATE','DEPTNO']);
print(emp_data)
```

```python
# using missing_values and filling_values argument
```

```python
emp_data=np.genfromtxt(file_name, dtype=data_type, comments='\\', names=True, missing_values={'COMM':'N/A'});
print(emp_data)
```

```python
emp_data=np.genfromtxt(file_name, dtype=data_type, comments='\\', names=True, missing_values={'COMM':'N/A'}, filling_values={'COMM':0});
print(emp_data)
```

```python
# the converters argument
```

```python
from datetime import datetime;
data_type=(int,'|U8','|U10',int,'datetime64[D]',float,int,int)
date_fmt="%d-%m-%Y";
convert=lambda x:  datetime.strptime(str(x),date_fmt).date();
emp_data=np.genfromtxt(file_name, dtype=data_type, comments='\\', names=True, converters={'HIREDATE': convert});
print(emp_data)
```
