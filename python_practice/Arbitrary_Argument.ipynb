### Arbitrary Argument

```python
def multiply(*number):
    product=1;
    for i in number:
        product*=i;
    return product;
```

```python
multiply(1,2,3,4)
```

```python
multiply(34,43)
```

```python
multiply(34)
```

```python
def Print(*value):
    for i in value:
        print(i,end=' ');
    print();
```

```python
Print('abbas',43,'kalam');
```

```python
Print([34,34,623])
```

```python
def ftn():
    x=10;
    print("ID of x : ",id(x));
    print("value of x : ",x);
    y=x;
    print("ID of y : ",id(y));
    print("value of y : ",y);
    y=4;
    print("ID of y after changing the value : ",id(y));
    print("value of y : ",y);
```

```python
def ftn():
    global x
    x=5;
    print("ID of x : ",id(x));
    print("value of x : ",x);
    y=x;
    print("ID of y : ",id(y));
    print("value of y : ",y);
```

```python
x=3;
print("ID of x outside the function : ",id(x));
print("value of x outside the function : ",x);
print();
ftn();
print("\nID of x after function call : ",id(x));
print("value of x after function call : ",x);
```
