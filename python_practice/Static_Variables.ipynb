### Staic or Class Variables

```python
class Customer:
    
    __counter=0;            # private and static Variable
    def __init__(self):
        Customer.__counter+=1;
        self.customer_id=Customer.__counter;
        self.name=None;
        self.gender=None;
        self.age=None;
    
    def add_info(self,name,gender,age):
        self.name=name;
        self.gender=gender;
        self.age=age
        
    def show_info(self):
        print("Customer ID : ",self.customer_id);
        print("Name : ",self.name);
        print("Gender : ",self.gender);
        print("Age : ",self.age);
        print();
```

```python
C1=Customer();
```

```python
C2=Customer();
```

```python
C3=Customer();
```

```python
C4=Customer();
```

```python
C1.add_info('khalid','male',23);
```

```python
C2.add_info("Usman",'male',20);
```

```python
C3.add_info("Laiba",'Female',23);
```

```python
C4.add_info("wahab Riaz",'male',40);
```

```python
customers=[C1,C2,C3,C4];
```

```python
for customer in customers:
    customer.show_info();
```
