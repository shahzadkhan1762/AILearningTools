### Adding, removing, and sorting elements

***This section covers np.sort(), np.concatenate()***

```python
import numpy as np;
```

```python
arr=np.array([[4,36,28,39,73,93]]);
```

```python
print(arr);
```

```python
np.sort(arr);
```

```python
print(arr);
```

```python
arr2=np.array([[43,26,25]]);
```

```python
np.concatenate(arr,arr2);
```

```python
arr1=np.array([5,2,73,26]);
arr2=np.array([29,69,26]);
```

```python
np.concatenate((arr1,arr2))
```

```python
print(arr1)
```

```python
print(arr2);
```

```python
np.concatenate((arr1,arr2))
```
