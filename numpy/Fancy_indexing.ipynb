### Indexing using boolean arrays

```python
arr1=np.random.randint(low=0,high=100,size=20).reshape(4,5);
```

```python
arr1
```

```python
arr1>40
```

```python
arr1[arr1>40]
```

```python
arr1[(arr1>40) & (arr1%2==0)]
```

```python
arr1[(arr1>30) & (arr1%2==0)]=0
```
